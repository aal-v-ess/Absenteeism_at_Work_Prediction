## loop to insert data into database

```python
insert_query = 'INSERT INTO database_name VALUES '
```

```python
insert_query
```

```python
for i in range(df_new_obs.shape[0]):
    insert_query += '('
    
    for j in range(df_new_obs.shape[1]):
        insert_query += str(df_new_obs[df_new_obs.columns.values[j]][i]) + ', '
        
    insert_query = insert_query[:-2] + '), '
```

```python
insert_query
```

```python
insert_query = insert_query[:-2] + ';' # reverse index to corret end of statement in mysql
insert_query
```

```python
cursor.execute(insert_query)
```

```python
conn.commit()
```

```python
conn.close()
```
