## Analyse Reason for absense column

```python
df['Reason for absence'].unique()
```

```python
# Reason for absense is unique for each user and the description in the data description file
print(sorted(df['Reason for absence'].unique())) # missing number 20
```

```python
# check if a user had more than one reason for absense (result is only one so each user gave only one reason for absense)
pd.get_dummies(df['Reason for absence']).sum(axis=1).unique()
```

```python
reason_columns = pd.get_dummies(df['Reason for absence'], drop_first=True)
reason_columns
```

```python
# Avoid multicollinearity
df.drop(columns=['Reason for absence'], inplace=True)
df.head()
```

```python
# Group absense reasons by type of reason
reason_type_1 = reason_columns.loc[:, 1:14].max(axis=1) # diseases
reason_type_2 = reason_columns.loc[:, 15:17].max(axis=1) # Injuries
reason_type_3 = reason_columns.loc[:, 18:21].max(axis=1) # Pregnancy related
reason_type_4 = reason_columns.loc[:, 22:28].max(axis=1) # routine stuff
# reason_type_4 = reason_columns.loc[:, 22;].max(axis=1)
```

## Concatenate reason for absense column to df

```python
df = pd.concat([df, reason_type_1, reason_type_3, reason_type_3, reason_type_4], axis=1)
```

```python
df.rename(columns={0:'Reason_1', 1:'Reason_2', 2:'Reason_3', 3:'Reason_4'}, inplace=True)
df.head()
```
